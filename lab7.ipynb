Task 1

```python
import pygame # type: ignore
import datetime

pygame.init()

WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))

img1 = pygame.image.load("clock.png")  
img2 = pygame.image.load("rightarm.png")
img3 = pygame.image.load("leftarm.png")

img1 = pygame.transform.scale_by(img1, 800 / 1400)
img2 = pygame.transform.scale_by(img2, 800 / 1400)
img3 = pygame.transform.scale_by(img3, 800 / 1400)

running = True
clock = pygame.time.Clock()

while running:
    screen.fill((255, 255, 255))
    screen.blit(img1, (0, 0))

    now = datetime.datetime.now()
    minutes = now.minute
    seconds = now.second

    angle2 = -(minutes * 6) - 54
    angle1 = -(seconds * 6) - 4
    
    rotated_img1 = pygame.transform.rotate(img3, angle1)
    rotated_img2 = pygame.transform.rotate(img2, angle2)

    rect1 = rotated_img1.get_rect(center=(WIDTH // 2, HEIGHT // 2))
    screen.blit(rotated_img1, rect1.topleft)
    rect2 = rotated_img2.get_rect(center=(WIDTH // 2, HEIGHT // 2))
    screen.blit(rotated_img2, rect2.topleft)

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    pygame.display.update()
    clock.tick(1)

pygame.quit()
```

Task 2

```python
import pygame # type: ignore
import os

pygame.init()
pygame.mixer.init()

WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))

music_folder = "music"

songs = [f for f in os.listdir(music_folder)]
current_track = 0
is_playing = False

def play_music():
    global is_playing
    pygame.mixer.music.load(os.path.join(music_folder, songs[current_track]))
    pygame.mixer.music.play()
    is_playing = True

def toggle_play_pause():
    global is_playing
    if pygame.mixer.music.get_busy():
        pygame.mixer.music.pause()
        is_playing = False
    else:
        pygame.mixer.music.unpause()
        is_playing = True

def next_track():
    global current_track
    current_track = (current_track + 1) % len(songs)
    play_music()

def previous_track():
    global current_track
    current_track = (current_track - 1) % len(songs)
    play_music()

running = True

if songs:
    play_music()

# SPACE - PAUSE/PLAY 
# RIGHT ARROW - NEXT TRACK 
# LEFT ARROW - PREVIOUS TRACK

while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_SPACE:
                toggle_play_pause()
            elif event.key == pygame.K_RIGHT:
                next_track()
            elif event.key == pygame.K_LEFT:
                previous_track()

pygame.quit()
```

```python
import pygame # type: ignore

pygame.init()

WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))

radius = 25
x = WIDTH // 2
y = HEIGHT // 2

color = (255, 0, 0)
bg_color = (255, 255, 255)

step = 20

running = True
while running:
    screen.fill(bg_color)
    pygame.draw.circle(screen, color, (x, y), radius)

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_UP and y - radius - step >= 0:
                y -= step
            elif event.key == pygame.K_DOWN and y + radius + step <= HEIGHT:
                y += step
            elif event.key == pygame.K_RIGHT and x + radius + step <= WIDTH:
                x += step
            elif event.key == pygame.K_LEFT and x - radius - step >= 0:
                x -= step
    
    pygame.display.update()

pygame.quit()
```
